```python
from torchvision.models import get_model
import torch as th
import torch.nn as nn
```

```python
# Русский: загружаю стандартную предобученную модель ResNet-18 из torchvision, чтобы посмотреть её архитектуру
# English: loading the standard pretrained ResNet-18 model from torchvision to inspect its architecture

resnet18 = get_model("resnet18", weights="DEFAULT")
resnet18
```

```python
resnet18 = get_model("resnet18", weights="DEFAULT")
resnet18.fc = nn.Identity()
th.save(resnet18, "resnet18_extractor.pth")
sum([p.numel() for p in resnet18.parameters()])
```
